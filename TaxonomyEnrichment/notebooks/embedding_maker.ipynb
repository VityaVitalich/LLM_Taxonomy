```python
from gensim.models import KeyedVectors
import pickle
```

```python
path = '../TaxoEnrich/data/MAG_CS/new/test/computer_science.terms.embed'
path = '/home/LLM_Taxonomy/TaxonomyEnrichment/data/verb/embeddings/embeddings_ft.pickle'
```

```python
with open(path, 'rb') as f:
    emb = pickle.load(f)
```

```python
test_path = '../data/verb/verb.terms'
with open(test_path, 'r') as f:
    ls = f.readlines()

test = {}
inv_test = {}
for item in ls:
    idx, term = item.replace('\n', '').split('\t')
   # idx = int(idx)
    test[idx] = term
    inv_test[term] = idx
```

```python
res_str = ""
for idx, term in test.items():
    cur_embedding = ' '.join(list(emb[term].astype(str)))
    cur_str = idx + " " + cur_embedding

    res_str += cur_str + '\n'
```

```python
res_str[:100]
```

```python
final_str = str(len(test)) + ' ' + '300\n' + res_str
```

```python
with open('../data/verb/verb.terms.embed', 'w') as f:
    f.write(final_str)

with open('../data/verb/verb.terms.bertembed', 'w') as f:
    f.write(final_str)
```
