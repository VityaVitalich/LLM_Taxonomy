```python
import networkx as nx
import pickle
from tqdm import tqdm
from copy import deepcopy
```

```python
train = nx.read_edgelist('../data/MAG_CS/train.edgelist', create_using=nx.DiGraph, delimiter='\t')
```

```python
with open('/home/LLM_Taxonomy/TaxonomyEnrichment/data/MAG_CS/test_hypernyms_def.pickle', 'rb') as f:
    test_hypernyms = pickle.load(f)
```

```python
test_hypernyms
```

```python
len(train.nodes())
```

```python
all_pairs = []

for cur_elem in tqdm(test_hypernyms):
    elem = deepcopy(cur_elem)
    all_pairs.append(elem)
    for node in train.nodes():
        elem = deepcopy(cur_elem)
        elem['parents'] = node
        all_pairs.append(elem)
```

```python
codes = []
for elem in all_pairs:
    codes.append(elem['children'] + elem['parents'])
```

```python
len(codes)
```

```python
len(set(codes))
```

```python
with open('../data/MAG_CS/all_pairs.pickle', 'wb') as f:
    pickle.dump(all_pairs, f)
```
