```python
import networkx as nx
import pickle
from tqdm import tqdm
from copy import deepcopy
```

```python
train = nx.read_edgelist('../data/psychology/train.edgelist', create_using=nx.DiGraph, delimiter='\t')
```

```python
with open('/home/LLM_Taxonomy/TaxonomyEnrichment/data/psychology/test_hypernyms_def.pickle', 'rb') as f:
    test_hypernyms = pickle.load(f)
```

```python
test_hypernyms
```

```python
len(train.nodes())
```

```python
all_pairs = []

for cur_elem in tqdm(test_hypernyms):
    elem = deepcopy(cur_elem)
    if elem['case'] == 'predict_hypernym':
        all_pairs.append(elem)
    else:
        for parent in elem['parents']:
            elem = deepcopy(cur_elem)
            elem['parents'] = parent
            elem['case'] = 'predict_hypernym'
            all_pairs.append(elem)
    for node in train.nodes():
        elem = deepcopy(cur_elem)
        elem['parents'] = node
        elem['case'] ='predict_hypernym'
        all_pairs.append(elem)

                
```

```python
all_pairs[1]
```

```python
codes = []
for elem in all_pairs:
    codes.append(elem['children'] + elem['parents'])
```

```python
for elem in all_pairs:
    if isinstance(elem['parents'], list):
        print(1)
    if elem['case'] != 'predict_hypernym':
        print(1)
```

```python
with open('../data/psychology/all_pairs.pickle', 'wb') as f:
    pickle.dump(all_pairs, f)
```
