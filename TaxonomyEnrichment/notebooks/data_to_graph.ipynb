```python
train_path = '../data/noun/wordnet_noun.terms'
with open(train_path, 'r') as f:
    ls = f.readlines()

train = {}
for item in ls:
    idx, term = item.replace('\n', '').split('\t')
    train[idx] = term
```

```python
taxo_path = '../data/noun/wordnet_noun.taxo'
with open(taxo_path, 'r') as f:
    ls = f.readlines()

taxo = []
for item in ls:
    parent, child = item.replace('\n', '').split('\t')
    taxo.append((parent, child))
```

```python
G = nx.DiGraph()

for parent, child in taxo:
    parent_name = train[parent]
    child_name = train[child]

    G.add_node(parent_name)
    G.add_node(child_name)
    G.add_edge(parent_name, child_name)
```
