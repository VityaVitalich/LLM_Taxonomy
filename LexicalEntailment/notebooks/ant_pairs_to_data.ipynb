```python
all_pairs_formated = []


for parent, pdef, child, cdef in test:
    elem = {}
    elem["children"] = child
    elem["grandparents"] = None
    elem["parents"] = parent
    elem["case"] = "predict_hypernym"
    elem['child_def'] = cdef

    all_pairs_formated.append(elem)

    elem = {}
    elem["children"] = parent
    elem["grandparents"] = None
    elem["parents"] = child
    elem["case"] = "predict_hypernym"
    elem['child_def'] = pdef

    all_pairs_formated.append(elem)
```
